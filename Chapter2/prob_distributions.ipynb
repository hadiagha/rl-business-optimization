```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm

# Parameters
lam = 4
mu = lam
sigma = np.sqrt(lam)

# Define support for distributions
x_p = np.arange(poisson.ppf(0.01, lam), poisson.ppf(0.99, lam) + 1)
x_n = np.linspace(norm.ppf(0.01, mu, sigma), norm.ppf(0.99, mu, sigma), 200)

# Compute PMF and PDF
pmf_p = poisson.pmf(x_p, lam)
pdf_n = norm.pdf(x_n, mu, sigma)

# Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
plt.rcParams.update({
    "ytick.labelsize": 13,
    "xtick.labelsize": 13
})

# Poisson distribution plot
axes[0].bar(x_p, pmf_p)
axes[0].set_title(f'Poisson Distribution (λ={lam})', fontsize=14)
axes[0].set_xlabel('k', fontsize=14)
axes[0].set_ylabel('Probability Mass', fontsize=14)
axes[0].grid(False)

# Normal distribution plot
axes[1].plot(x_n, pdf_n)
axes[1].set_title(f'Normal Distribution (μ={mu}, σ={sigma:.2f})', fontsize=14)
axes[1].set_xlabel('x', fontsize=14)
axes[1].set_ylabel('Probability Density', fontsize=14)
axes[1].grid(False)

# Adjust layout and display
fig.tight_layout()
plt.show()
```
